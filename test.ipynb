### Test plotting

```python
import matplotlib.pyplot as plt
import numpy as np

nrows = 3
ncols = [5,1,1]
maxcols = max(ncols)
fig, axes = plt.subplots(nrows=nrows, ncols=maxcols, figsize=(80, 10), dpi = 250)

data_array = np.random.rand(3, 10, 128, 128, 2)
save_path = 'test_image.png'


for row in range(axes.shape[0]):
    for col in range(ncols[row]):
        ax = axes[row, col]

        print(row, col)
        img = ax.imshow(data_array[row][col][:,:,0], cmap='RdBu_r')

fig.colorbar(img, ax=axes.ravel().tolist(), orientation='vertical', fraction=0.015, pad=0.04)

# print("label while plotting:", label)
plt.savefig(save_path)
```
